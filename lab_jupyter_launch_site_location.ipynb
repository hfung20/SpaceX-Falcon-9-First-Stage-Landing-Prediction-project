```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

# Helper function to calculate distance in KM
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0 # Radius of earth
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon, dlat = lon2 - lon1, lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    return R * (2 * atan2(sqrt(a), sqrt(1 - a)))

# Example: Distance from VAFB SLC-4E (34.63283, -120.61074) to Coastline (34.634, -120.625)
launch_site_lat, launch_site_lon = 34.63283, -120.61074
coast_lat, coast_lon = 34.636, -120.624
distance = calculate_distance(launch_site_lat, launch_site_lon, coast_lat, coast_lon)

# Add a marker with the distance text
distance_marker = folium.Marker(
   [coast_lat, coast_lon],
   icon=DivIcon(
       icon_size=(20,20),
       icon_anchor=(0,0),
       html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
   )
)
site_map.add_child(distance_marker)

# Draw the line (PolyLine)
lines = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [coast_lat, coast_lon]], weight=1)
site_map.add_child(lines)

site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
